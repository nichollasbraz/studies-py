## Learning check 7
Escreva uma função chamada average(), que recebe como entrada uma lista de valores numéricos. A função deve retornar a média dos valores na lista. Se a lista for vazia, a função deve retornar o valor 0.


```python
def average(lista):
  if len(lista) == 0:
    return 0
  else:
    return sum(lista)/len(lista)

def average2(lista):
  if len(lista) == 0:
    return 0
  else:
    soma=0
    for item in lista:
      soma=soma+item # soma+=item
    return soma/len(lista)


print(average([1,2,3]))
print(average2([1,2,3]))
```

## Exercício 7
Uma função em jogos é a detecção de colisão entre objetos. Escreva uma função chamada colisão() que verifica e retorna True se dois objetos circulares colidem e retorna False, caso contrário. Cada objeto circular será dado por seu raio e as (x, y) coordenadas de seu centro. Assim, a função tomará seis números como entrada: as coordenadas x1 e y1 do centro e o raio r1 do primeiro círculo, e as coordenadas x2 e y2 do centro e o raio r2 do segundo círculo.

<img src="colisao.gif" width="500">

```python
import math as mt
def colisao(x1,y1,r1,x2,y2,r2):
    d=mt.sqrt((x1-x2)**2+(y1-y2)**2)
    if d<=(r1+r2):
        return True
    else:
        return False

x1,y1,r1,x2,y2,r2=5,2,2,0,0,2
resultado=colisao(x1,y1,r1,x2,y2,r2)
# print(resultado)
print(f'O resultado da função que testa se vai haver colisão é {resultado}')
```

## Learning check 9
Escreva uma função chamada intersect(), que recebe duas listas como parâmetro. A função deve retornar uma lista contendo todos os itens que estão em ambas as listas. A lista resultante deve estar ordenada.

```python
def intersect(lista1,lista2):
  nasDuas = []
  for elemento in lista1:
    if elemento in lista2 and elemento not in nasDuas:
      nasDuas.append(elemento)
  nasDuas.sort()
  return nasDuas

print(intersect([3,4,4,5],[6,7,8,4]))
print(intersect([4, 76, -1, -2], [19, -2, -1, 27, 4, -3]))
```

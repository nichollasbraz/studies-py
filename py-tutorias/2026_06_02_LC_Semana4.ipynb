### Learning Check 1
#### Objetivo
Neste problema, praticaremos o mecanismo de definição de funções e o conceito de saída formatada.  

#### Especificação
Esta questão exige a implementação de duas funções e, por isso, é dividida nas duas partes a seguir. 

##### Primeira parte

Nos Estados Unidos, a temperatura é medida em graus Fahrenheit em vez de graus Celsius. É bem simples converter uma esacla na outra: se **C** representa a temperatura em Celsius e **F**, a tempratura em Fahrenheit correspondente é dada por: 

$ F = (9C/5)+32 $.

Escreva uma função chamada `convert()` que recebe como parâmetro um número que representa uma temperatura em Celsius e retorna o correspondente em Fahrenheit. A sua função deve-se comportar como a seguir:

```
>>> convert(0)
32.0
>>> convert(20)
68.0
```

##### Segunda parte

Você deve definir uma função chamada `table()` que, usando a função definida na primeira parte, imprime uma tabela formatada que ajuda um turista vindo do Brasil a converter as temperaturas entre as diferentes escalas. A sua tabela deve mostrar as temperaturas entre -30 e 40 (variando de 10 em 10 graus) Celsius e o seu correspondente em Fahrenheit, como a seguir:
```
>>> table()
  F       C
-22.0   -30.0
 -4.0   -20.0
 14.0   -10.0
 32.0     0.0
 50.0    10.0
 68.0    20.0
 86.0    30.0
104.0    40.0
```

### Learning Check 6
#### Objetivo
Neste problema, praticaremos a definição de funções e leitura de arquivos.  

#### Especificação

Baixamos um arquivo chamado `sobrenomesAmericanos.txt*` do site do censo americano (http://www.census.gov/). Nesse arquivo, cada linha corresponde a um sobrenome (com letras maiúsculas) e um número correspondente ao percentual da população americana que tem o determinado sobrenome. O começo do arquivo é mostrado a seguir: 

```
SMITH	0.88085
JOHNSON	0.68844
WILLIAMS	0.56866
BROWN	0.51162
...
```

Por exemplo, 0.88085% dos americanos têm SMITH como sobrenomes. Escreva uma função chamada `freq()` que use os dados no arquivo `sobrenomesAmericanos.txt` para computar e retornar uma aproximação para o número de americanos que têm um dado sobrenome.  Você pode assumir que a população americana atual é, aproximadamente, 300000000 (= 3*10**8). Um exemplo de como a sua função deve funcionar está apresentado a seguir:

```
>>> freq('Smith')
264255000
>>> freq('Ferreira')
0
```

Note que, se o nome procurado não estiver na lista, a sua função deve retornar 0. A sua função não deve ser `case-sensitive`, ou seja, não há diferença entre letras maiúsculas e minúsculas.



### Learning Check 7

Neste problema, praticaremos a definição de funções e leitura de arquivos.  

#### Especificação

Em um arquivo chamado `currencies.txt`, incluímos a conversão de moedas de diversos países do mundo para dólares americanos. O começo do arquivo é mostrado a seguir: 

```
AUD	1.0345157	Australian Dollar
CHF	1.0237414	Swiss Franc
CNY	0.1550176	Chinese Yuan
DKK	0.1651442	Danish Krone
EUR	1.2296544	Euro
GBP	1.5550989	British Pound
HKD	0.1270207	Hong Kong Dollar
INR	0.0177643	Indian Rupee
JPY	0.01241401	Japanese Yen
MXN	0.0751848	Mexican Peso
MYR	0.3145411	Malaysian Ringgit
NOK	0.1677063	Norwegian Krone
NZD	0.8003591	New Zealand Dollar
PHP	0.0233234	Philippine Peso
SEK	0.148269	Swedish Krona
SGD	0.788871	Singapore Dollar
THB	0.0313789	Thai Baht
...
```

Por exemplo, 'JPY' é a sigla para o yen japonês, e 'EUR' para o euro. Então, de acordo com a informação acima, 1 euro equivale a 1.2296544 dólares americanos.

Escreva uma função chamada `curconv()` que recebe como entradas uma string especificando uma moeda (por meio da sua abreviação) e uma quantia e retorna o valor dessa quantia convertido para dólares americanos, usando a informação no arquivo `currencies.txt`.
Um exemplo de como a sua função deve funcionar está apresentado a seguir:

```
>>> curconv('EUR', 100)
122.96544
>>> curconv('JPY', 100)
1.241401
```

Caso a string que especifica a moeda fornecida não esteja contida no arquivo, o função deve retornar -1.


### Learning Check 8
#### Objetivo
Neste problema, praticaremos a definição de funções, a leitura e a escrita de arquivos.  

#### Especificação

Muitos websites produzem nuvens de palavras (https://www.wordclouds.com/), mostrando quais são os tópicos mais discutidos ou quais palavras foram usadas em um texto. O primeiro passo para produzir tais "tags clouds" é remover as chamadas *stop words* do texto em questão. Estas são palavras comuns, como "de", "que", "para", "tem", "mais", entre outras, que não expressam um tópico definido e, portanto, são ignoradas para demostrar o sentido do texto em nuvens de palavras.

Escreva uma função chamada `clean()` que recebe como parâmetro o nome de um arquivo de texto. Essa função então lê o texto nesse arquivo, divide esse texto em palavras e remove as palavras listadas no arquivo `stopwords.txt` e, finalmente, produz um novo arquivo chamado `words.txt`, que contém as palavras remanescentes. Nesse novo arquivo, cada palavra aparece na ordem de aparacição do texto original, cada uma em uma linha diferente e em letras minúsculas.

## Criação de um jogo usando módulos

**Adivinhe o número**

Computador sorteia um número e o usuário tenta acertá-lo.

### Primeira solução: básico

````python
import random as rd


def sorteia_numero(inicio, fim):
    return rd.randint(inicio, fim)


def le_palpite():
    while True:
        try:
            palpite = int(input('Digite seu palpite: '))
            return palpite
        except ValueError:
            print('Entrada inválida. Digite um número inteiro.')


def mostra_dica(numero_secreto, palpite):
    if palpite < numero_secreto:
        print('O número é maior.')
    elif palpite > numero_secreto:
        print('O número é menor.')
    else:
        print('Parabéns! Você acertou!')


print('Bem-vindo ao jogo de Adivinhação!')
print('Tente adivinhar o número entre 1 e 100.\n')

numero_secreto = sorteia_numero(1, 100)
tentativas = 0

while True:
    palpite = le_palpite()
    tentativas = tentativas + 1

    mostra_dica(numero_secreto, palpite)

    if palpite == numero_secreto:
        print(f'Você acertou em {tentativas} tentativa(s)!')
        break

print('\nObrigado por jogar. Até a próxima!')
````

### Segunda solução: separação em módulos:

O jogo vai ser separado em módulos específicos, para organização do código: 
- Principal: `principal.py`
  - Controla o fluxo do jogo e chama os demais módulos
- Exibição: `exibição.py`
  - Cuida de todas as mensagens exibidas para o jogador (interface textual)
- Front-end: `frontend.py`
  - Gerencia a lógica de adivinhação, sorteio do número, verificação dos palpites, interação com o jogador


````python
# exibicao.py
def boas_vindas():
    print("Bem-vindo ao jogo de Adivinhação!")
    print("Tente adivinhar o número entre 1 e 100.\n")

def dica(numero, palpite):
    if palpite < numero:
        print("O número é maior.")
    elif palpite > numero:
        print("O número é menor.")
    else:
        print("Parabéns! Você acertou!")

def despedida():
    print("\n👋 Obrigado por jogar. Até a próxima!")

def tentativa_invalida():
    print("Entrada inválida. Digite um número inteiro.")
````

````python
# frontend.py
import random
import exibicao

def jogar():
    numero_secreto = random.randint(1, 100)
    tentativas = 0

    while True:
        try:
            palpite = int(input("Digite seu palpite: "))
        except ValueError:
            exibicao.tentativa_invalida()
            continue

        tentativas += 1
        exibicao.dica(numero_secreto, palpite)

        if palpite == numero_secreto:
            print(f"Você acertou em {tentativas} tentativa(s)!")
            break
 
````

````python
# principal.py
import exibicao
import frontend 

def main():
    exibicao.boas_vindas()
    frontend.jogar()
    exibicao.despedida()

if __name__ == "__main__": #aqui, testa se está sendo executado diretamente ou se está sendo importado
    main()
````

### Terceira solução: uso de uma classe

* Até agora, o estado da partida era mantido por variáveis independentes
* Com a criação da classe, essas informações passam a pertencer ao objeto que representa a partida
* A lógica da partida foi deslocada para a classe `JogoAdivinhacao`
* O `frontend` passa a interagir com um objeto que controla o estado do jogo


````python
# jogo.py
import random

class JogoAdivinhacao:
    def __init__(self, minimo=1, maximo=100):
        self.minimo = minimo
        self.maximo = maximo
        self.numero_secreto = random.randint(minimo, maximo)
        self.tentativas = 0
        self.acertou = False

    def registra_palpite(self, palpite):
        self.tentativas = self.tentativas + 1

        if palpite < self.numero_secreto:
            return 'maior'
        elif palpite > self.numero_secreto:
            return 'menor'
        else:
            self.acertou = True
            return 'acertou'
````

````python
# exibicao.py
def boas_vindas():
    print('Bem-vindo ao jogo de Adivinhação!')
    print('Tente adivinhar o número entre 1 e 100.\n')

def dica(resultado):
    if resultado == 'maior':
        print('O número é maior.')
    elif resultado == 'menor':
        print('O número é menor.')
    else:
        print('Parabéns! Você acertou!')

def despedida():
    print('\n👋 Obrigado por jogar. Até a próxima!')

def tentativa_invalida():
    print('Entrada inválida. Digite um número inteiro.')
````

````python
# frontend.py
import exibicao
from jogo import JogoAdivinhacao

def jogar():
    jogo = JogoAdivinhacao(1, 100)

    while not jogo.acertou:
        try:
            palpite = int(input('Digite seu palpite: '))
        except ValueError:
            exibicao.tentativa_invalida()
            continue

        resultado = jogo.registra_palpite(palpite)
        exibicao.dica(resultado)

    print(f'Você acertou em {jogo.tentativas} tentativa(s)!')
````

````python
# principal.py
import exibicao
import frontend 

def main():
    exibicao.boas_vindas()
    frontend.jogar()
    exibicao.despedida()

if __name__ == '__main__': #aqui, testa se está sendo executado diretamente ou se está sendo importado
    main()
````

## Learning Check 5
Escreva uma classe chamada `Selection` com um construtor que recebe uma lista como entrada. A classe deve ter um método chamado `next()` que retorna o próximo item da lista, de forma contínua.

Por exemplo:
```text
>>> lst = ['Zoe', 'Yannick', 'Xena', 'Wendy']
>>> sel = Selection(lst)
>>> sel.next()
'Zoe'
>>> sel.next()
'Yannick'
>>> sel.next()
'Xena'
>>> sel.next()
'Wendy'
>>> sel.next()
'Zoe'
```
Dessa maneira, ao chegar ao final da lista, você deve retornar o primeiro elemento, quando o método `next()` for chamado. 

````python
class Selection:
    def __init__(self,lst):
        self.lst = lst
        self.index = -1

    def next(self):
        self.index = (self.index + 1) % len(self.lst)
        return self.lst[self.index]
````

## Exercício 1
Criar a classe `FIGURA` para representar figuras geométricas:
* Essa classe deve ter classes “filhas”:
    * Quadrado
    * Retângulo
    * Círculo

* Essas classes devem ter métodos apropriados para cálculo de área e perímetro


````python
class Figura:
    def area(self):
        return 'area não definida'
class Quadrado(Figura):
    def __init__(self,lado=0):
        self.lado=lado
    def area(self):
        return self.lado**2
    def perimetro(self):
        return self.lado*4
class Retangulo(Figura):
    def __init__(self,largura=0,altura=0):
        self.largura=largura
        self.altura=altura
    def area(self):
        return self.largura*self.altura
    def perimetro(self):
        return (self.largura+self.altura)*2
class Circulo(Figura):
    def __init__(self,raio=0):
        self.raio=raio
    def area(self):
        import math as mt
        return mt.pi*self.raio**2
    def perimetro(self):
        import math as mt
        return mt.pi*self.raio*2
class Hexagono(Figura):
    def __init__(self,lado=0):
        self.lado=lado
    def perimetro(self):
        return self.lado*6
````

## Exercício 2
Vamos alterar um pouco mais a classe `Fila`. Agora, ela deve:
* Informar o tamanho usando uma variação do método len
* Dar retorno de uma representação da instância

Essa classe deve ter uma classe filha que permita desenfileirar usando o método LIFO

````python
class Fila:
    'uma classe de fila clássica'

    def __init__(self,valores=''):
        'instancia uma lista vazia'
        if valores=='':
            self.q = []
        else:
            if type(valores)==list:
                self.q = valores
            else:
                self.q = list(eval(valores))
                

    def vazia(self):
        'retorna True se a fila estiver vazia, False caso contrário'
        return (len(self.q) == 0)

    def enfilera (self, item):
        'Insere item no final da fila'
        return self.q.append(item)

    def desenfilera(self):
        'remove e retorna item na frente da fila '
        if len(self.q)==0:
            return False
        return self.q.pop(0)
    
    def tamanho(self):
        'retorna o tamanho da fila'
        return len(self.q)
    def __repr__(self): 
        return f"Fila({self.q})"
    def __len__(self):
        return len(self.q)
        
    
class FilaOpt(Fila):
    'uma classe de fila com opção de método de saída'
    def desenfilera(self,metodo='FIFO'):
        'Remove um item da fila, de acordo com o método indicado (FIFO ou LIFO)'
        if len(self.q)==0:
            return False
        if metodo=='LIFO':
            return self.q.pop()
        else:
            return self.q.pop(0)
````

````python
class SaldoInsuficienteException(Exception):
    def __init__(self, saldo, valor_saque):
        super().__init__(f'Saldo insuficiente: saldo atual = R${saldo:.2f}, saque solicitado = R${valor_saque:.2f}')

class ContaBancaria:
    def __init__(self, agencia, conta, saldo=0):
        self.agencia = agencia
        self.conta = conta
        self.saldo = saldo

    def setAgencia(self, nova_agencia):
        'Altera o número da agência'
        self.agencia = nova_agencia

    def setConta(self, nova_conta):
        'Altera o número da conta'
        self.conta = nova_conta

    def setSaldo(self, novo_saldo):
        'Altera o saldo da conta'
        self.saldo = novo_saldo

    def saque(self, valor):
        'Realiza um saque, se houver saldo suficiente, senão lança exceção'
        if valor <= self.saldo:
            self.saldo -= valor
            return True  # saque realizado
        else:
            raise SaldoInsuficienteException(self.saldo, valor)

    def deposito(self, valor):
        'Realiza um depósito na conta'
        self.saldo += valor

    def get(self):
        'Retorna uma tupla com (agência, conta, saldo)'
        return (self.agencia, self.conta, self.saldo)
    
    def __add__(self,valor):
        'Adiciona valor ao saldo'
        self.saldo+=valor
    
    def __str__(self):
        return f'Agência: {self.agencia}, Conta: {self.conta}, Saldo: R$ {self.saldo:.2f}'
    
class ContaPoupanca(ContaBancaria):
    def rendimento(self):
        self.saldo *= 1.05  # aumenta o saldo em 5%
````

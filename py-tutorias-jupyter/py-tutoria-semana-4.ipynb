## Exercício
Criar funções para:
* Contar o número de caracteres dentro do arquivo
* Contar o número de linhas do arquivo
* Contar o número de palavras do arquivo

```python
def numChars(filename):
    'retorna o número de caracteres em um arquivo'

    infile = open(filename, 'r')
    content = infile.read()
    infile.close()

    return len(content)
print(numChars('existe.txt'))
```

```python
def numLines(filename):
    'retorna o número de linhas em um arquivo'

    infile = open(filename, 'r')
    lineList = infile.readlines()
    infile.close()

    return len(lineList)
print(numLines('existe.txt'))
```

```python
def numWords(filename):
    'retorna o número de palavras em um arquivo'

    infile = open(filename, 'r')
    content = infile.read()
    infile.close()
    wordList = content.split()

    return len(wordList)
print(numWords('existe.txt'))
```

## Exercício
Implemente a função `meuGrep()` que:
* Toma como entrada duas strings: um nome de arquivo e uma string alvo
* Exibe cada linha do arquivo que contém a string alvo



```python
def meuGrep(nome_arquivo,palavra):
    arquivo=open(nome_arquivo,'r',encoding='utf-8')
    linhas=arquivo.readlines()
    arquivo.close()
    for linha in linhas:
        if palavra in linha:
            print(linha)

meuGrep('existe.txt','acento')
```

```python
def meuGrep2(nome_arquivo,palavra):
    arquivo=open(nome_arquivo,'r',encoding='utf-8')
    linhas=arquivo.readlines()
    arquivo.close()
    filtradas=[]
    for linha in linhas:
        if palavra in linha:
            filtradas.append(linha)
    return filtradas

print(meuGrep2('existe.txt','acento'))
```

## Exercício
Implemente a função `mediaVendas()` que:
* Toma como entrada o nome do arquivo com movimentação de estoques
* Retorna a média de vendas dos produtos


```python
def mediaVendas(nome):
    arquivo=open(nome,'r',encoding='utf-8')
    arquivo.readline()
    dados=arquivo.readlines()
    arquivo.close()
    total_vendas=0
    for linha in dados:
        campos=linha.strip().split(',')
        total_vendas=total_vendas+int(campos[2])
    media=total_vendas/len(dados)
    return media
print(mediaVendas('dados\estoque2.txt'))
```

## Exercício
Implemente a função `fcopy()` que:
* Toma como entrada duas strings com os nomes de dois arquivos
* Copia o conteúdo do primeiro arquivo para o segundo


```python
def fcopy(origem,destino):
    arquivo_origem=open(origem)
    conteudo_origem=arquivo_origem.read()
    arquivo_origem.close()

    arquivo_destino=open(destino,'w')
    arquivo_destino.write(conteudo_origem)
    arquivo_destino.close()
fcopy('dados\estoque.txt','dados\copia.txt')
```

## Exercício
Implemente a função `censura()` que:
* Toma como entrada uma string com o nome do arquivo de entrada
* Substitui as palavras de 4 letras por '@@@@'
* Salva o resultado no arquivo `censurado.txt`


```python
def censura(nome):
    arquivo=open(nome)
    conteudo=arquivo.read()
    arquivo.close()

    arquivo_saida=open('censurado.txt','w')

    palavras=conteudo.split()

    for palavra in palavras:
        if len(palavra) == 4:
            arquivo_saida.write('@@@@ ')
        else:
            arquivo_saida.write(palavra + ' ')
    
    arquivo_saida.close()
censura('existe.txt')
```

```python
def censura2(nome):
    arquivo=open(nome)
    linhas=arquivo.readlines()
    arquivo.close()

    arquivo_saida=open('censurado2.txt','w')

    for linha in linhas:
        palavras=linha.split()
        for palavra in palavras:
            if len(palavra) == 4:
                arquivo_saida.write('@@@@ ')
            else:
                arquivo_saida.write(palavra + ' ')
        arquivo_saida.write('\n')
    
    arquivo_saida.close()
censura2('existe.txt')
```

## Exercício
Implemente a função `relatorioNotas()` que:
* Toma como entrada uma string COM o nome do arquivo de entrada
* esse arquivo de única linha terá notas em forma de letra, separadas por espaços
* Calcule a distribuição de notas
* Salve o resultado no arquivo `relatorio.txt`


```python
def relatorioNotas(nome):
    arquivo=open(nome,'r')
    dados=arquivo.read()
    arquivo.close()

    notas=dados.split()

    arquivo_saida=open(r'dados\relatorio.txt','w')

    notas_possiveis=['A','A+','A-','B','B+','B-','C','C+','C-','D','D+','D-','E','E+','E-','F',]

    for nota in notas_possiveis:
        quantidade=notas.count(nota)
        if quantidade>0:
            if quantidade == 1:
                arquivo_saida.write(f'1 aluno obteve nota {nota}\n')
            else:
                arquivo_saida.write(f'{quantidade} alunos obtiveram nota {nota}\n')

    arquivo_saida.close()
relatorioNotas(r'dados\notas.txt')
```

### Revisão de conceitos - arquivos

```python
## Abrir arquivo
# Arquivo não encontrado - cuidado com nome e caminho
arquivo=open('naoexiste.txt')
```

```python
# Acessando o conteúdo do arquivo
arquivo=open('existe.txt')
for linha in arquivo:
    print(linha)
```

```python
# Lendo o conteúdo do arquivo já lido
for novas in arquivo:
    print(novas)
```

```python
# Fechando o arquivo
arquivo.close()
```

```python
# Lendo o conteúdo de diversas maneiras
# Lendo tudo
arquivo=open('existe.txt')
tudo=arquivo.read()
print(tudo)
arquivo.close()
```

```python
# Lendo o conteúdo de diversas maneiras
# Lendo tudo com encoding
arquivo=open('existe.txt', encoding='utf-8')
tudo=arquivo.read()
print(tudo)
arquivo.close()
```

```python
# Lendo alguns caracteres
arquivo=open('existe.txt', encoding='utf-8')
pedaco1=arquivo.read(10)
print(pedaco1)
pedaco2=arquivo.read(5)
print(pedaco2)
resto=arquivo.read()
print(resto)
arquivo.close()
```

```python
# Lendo as linhas
arquivo=open('existe.txt', encoding='utf-8')
lista_linhas=arquivo.readlines()
print(lista_linhas)
arquivo.close()
```

```python
# Lendo cada linha
arquivo=open('existe.txt', encoding='utf-8')
linha=arquivo.readline()
print(linha)
linha=arquivo.readline()
print(linha)
linha=arquivo.readline()
print(linha)
arquivo.close()
```

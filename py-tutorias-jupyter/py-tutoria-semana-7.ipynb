## Exercício 3
**Estimando o valor de $\pi$**



|Figura         |Medida|Área        |
|:---------------:|:------:|:------:|
|Quadrado       |Lado: $2l$  |$4l^2$  |
|Círcunferência |Raio: $l$   |$\pi l^2$|

Se "jogarmos" $n$ dardos no quadrado, a razão entre a quantidade de dardos que vai cair dentro do círculo ($k$) e o total de dados jogados é aproximadamente proporcional à razão das áreas:

$\frac{k}{n}\approx \frac{\pi l^2}{4l^2}$

Então, podemos calcular o valor aproximado de $\pi$:

$\pi\approx\frac{k}{n}$

Criar uma função chamada `aproxPi()` que simula o lançamento de uma grande quantidade de dardos em um quadrado de lado 2 ($l=1$) e calcula o valor aproximado de $\pi$. Teste com diferentes quantidades de dardos e verifique que quanto mais dardos são lançados, mais próximo ao valor real chegaremos.

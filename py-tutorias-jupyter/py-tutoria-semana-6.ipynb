## Exercício 1
Implemente a função `meuIMC()` que:

* Aceita como entrada a altura de uma pessoa (em metros) e o peso (em quilos);
* Calcula o índice IMC, pela fórmula: $ IMC=\frac{peso}{altura^2} $

Exibir:

* Abaixo do peso, se $IMC \lt 18.5$
* Normal, se $18.5 \le IMC \lt 25$
* Sobrepeso, se $IMC \ge 25$

## Exercício 12

    * 1- entrar
    * 2- exibir tentativas

* Na **opção 1**, o usuário deverá digitar uma senha. Se esta senha for igual a frase secreta (pythonEhLegal), então o usuário entra no sistema e zera as tentativas. Caso contrário, o número de tentativas é incrementado. Note que o usuário só pode tentar no máximo 3 vezes.
* Na **opção 2**, o sistema exibe quantas tentativas foram feitas até o momento.
* Na **opção 3**, o sistema deve finalizar

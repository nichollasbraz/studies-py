## Exercício 2

Implemente um programa em Python que utilize a classe `Retangulo` definida abaixo para fazer o seguinte:

* Leia do usuário os valores da largura e da altura de um retângulo
* Crie um objeto `Retangulo` com esses valores
* Mostre na tela:
    - A área do retângulo
    - O perímetro do retângulo
    - Uma mensagem informando se ele é ou não um quadrado.

````python
class Retangulo:
    def __init__(self, largura, altura):
        self.largura = largura
        self.altura = altura

    def area(self):
        return self.largura * self.altura

    def perimetro(self):
        return 2 * (self.largura + self.altura)

    def eh_quadrado(self):
        return self.largura == self.altura
````

## Exercício 3

Utilizando a classe `Carro` definida abaixo, escreva um programa em Python que:

* Crie um carro do modelo "Uno" com consumo de 15 km/l
* Abasteça o carro com 30 litros
* Tente dirigir por 200 km. Informe se foi possível completar o trajeto
* Exiba o nível de combustível restante no tanque
* Tente dirigir por mais 300 km. Informe novamente o sucesso da viagem e o combustível restante.

````python
class Carro:
    def __init__(self, modelo, consumo):
        self.modelo = modelo         # modelo do carro (ex: "Fusca")
        self.consumo = consumo       # consumo em km por litro
        self.combustivel = 0.0       # litros no tanque

    def abastecer(self, litros):
        self.combustivel += litros

    def dirigir(self, distancia):
        gasto = distancia / self.consumo
        if gasto <= self.combustivel:
            self.combustivel -= gasto
            return True
        else:
            return False

    def nivel_combustivel(self):
        return self.combustivel

    def autonomia(self):
         return self.combustivel*self.consumo
````
